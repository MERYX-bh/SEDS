````python
from urllib.request import urlopen
from bs4 import BeautifulSoup as bs
import lxml

url = 'https://en.wikipedia.org/wiki/Data_science'
page = urlopen(url).read().decode('utf-8')

soup = bs(page)

links = soup.find_all('a')
all_link_text = []

all_link_text.extend([a.text for a in links])

text = ' '.join(all_link_text)
text
````

````python
import string
translator = str.maketrans('', '', string.punctuation + string.digits)
text = text.translate(translator)
````

````python
import nltk
tokens = nltk.word_tokenize(text)
````

````python
import nltk
nltk.download('stopwords')
````

````python
from nltk.corpus import stopwords
en_stopwords = stopwords.words('english')
en_stopwords = set(en_stopwords)
words = text.lower().split()
words = [w for w in words if w not in en_stopwords]
````

````python
cnt_fdist = nltk.FreqDist(words)
data = cnt_fdist.most_common(20)
data
````

````python
import requests as rq
all_text = [] 
list_links=['https://www.heavy.ai/learn/data-science',
           'https://en.wikipedia.org/wiki/Data_science',
        #   'https://www.ibm.com/cloud/learn/data-science-introduction',
           'https://deepai.org/machine-learning-glossary-and-terms/data-science']

for link in list_links:

    html = rq.get(link).text
    soup = bs(html)
    paragraph_text = soup.find_all('p')
    all_text.extend([p.text for p in paragraph_text])

    text = ' '.join(all_text)
````

````python
import string
translator = str.maketrans('', '', string.punctuation + string.digits)
text = text.translate(translator)
````

````python
import nltk
tokens = nltk.word_tokenize(text)
````

````python
import nltk
nltk.download('stopwords')
````

````python
from nltk.corpus import stopwords
en_stopwords = stopwords.words('english')
en_stopwords = set(en_stopwords)
words = text.lower().split()
words = [w for w in words if w not in en_stopwords]
````

````python
cnt_fdist = nltk.FreqDist(words)
cnt_fdist.most_common(20)
````

````python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
wordcloud = WordCloud(collocations=False).generate(' '.join(words))
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
````
